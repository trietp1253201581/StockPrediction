```python
import pandas as pd

df = pd.read_csv('analysis_last.csv')

df
```

```python
def color_column(col):
    if col.name == 'model':
        return
    return pd.Series(
        pd.Series(col, index=col.index).rank(pct=True),  # normalize
        index=col.index
    ).apply(lambda x: f'background-color: rgba(0, {int(255 * x)}, 255, 0.6)')

df.style.apply(color_column, axis=0)
```
